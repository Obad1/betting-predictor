```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder

def predict_correct_score(user_id, home_team, away_team, model, label_encoder, num_classes=25):
    """
    Predict the correct score for a given match using a neural network and return prediction confidence.

    Parameters:
    user_id (str): User's ID.
    home_team (str): Home team name.
    away_team (str): Away team name.
    model: Pre-trained neural network model.
    label_encoder: LabelEncoder for team names.
    num_classes (int): Number of possible scoreline classes.

    Returns:
    dict: Predicted scoreline and confidence percentage.
    """
    def safe_transform(le, team_name):
        """Safely encode team names using the LabelEncoder."""
        if team_name in le.classes_:
            return le.transform([team_name])[0]
        else:
            raise ValueError(f"Team '{team_name}' not recognized.")

    # One-Hot Encoding for team names
    encoder = OneHotEncoder(handle_unknown='ignore')
    one_hot_encoded_teams = encoder.fit_transform([[home_team], [away_team]]).toarray()

    # Safely encode team names
    try:
        encoded_home = safe_transform(label_encoder, home_team)
        encoded_away = safe_transform(label_encoder, away_team)
    except ValueError as e:
        return {"Error": str(e)}

    # Create input feature vector
    input_features = np.zeros((1, 130))
    input_features[0, :2] = [encoded_home, encoded_away]
    input_features[0, 2:2 + len(one_hot_encoded_teams.flatten())] = one_hot_encoded_teams.flatten()

    # Run model prediction
    prediction_output = model.predict(input_features)

    # Debugging: Print the model output
    print("Prediction Output:", prediction_output)

    # Access the 'scoreline' prediction
    correct_score_probs = prediction_output['scoreline'][0]

    # Get predicted score index and confidence
    predicted_score_index = np.argmax(correct_score_probs)
    confidence = correct_score_probs[predicted_score_index] * 100

    # Generate possible scorelines dynamically
    possible_scores = [f"{i}-{j}" for i in range(5) for j in range(5)]
    predicted_scoreline = (
        possible_scores[predicted_score_index]
        if predicted_score_index < len(possible_scores)
        else "Unknown"
    )

    return {
        "Predicted Scoreline": predicted_scoreline,
        "Confidence": f"{confidence:.2f}%"
    }
```
